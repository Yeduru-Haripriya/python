```python
# Question 3 : Develop a script that assigns grades to students based on their score.

def grade(score):
    if score >=90:
        return 'A'
    elif score >=80:
        return 'B'
    elif score >=70:
        return 'C'
    elif score >=60:
        return 'D'
    else:
        return 'F'


score=float(input('Enter your score: '))
print(f" Your grade acoording to score is {grade(score)}" )
```

```python
#Question 4 :  Write a Python script that asks for the number of items and the price per item,
# then calculates and prints the total cost with a 10% discount for 10 or more items, or a 20% discount for 20 or more items.

items=int(input('Enter the number of items: '))
price=[]*items
i=0
while i<items:
    rate=float(input(f'Enter the price of item  : {i} '))
    price.append(rate)
    i=i+1

if items>=10:
    total_cost=sum(price)*0.9
elif items>=20:
    total_cost=sum(price)*0.8
else:
    total_cost=sum(price)

print(f"Total cost is {total_cost} after applying any discounts if applicable")

```
